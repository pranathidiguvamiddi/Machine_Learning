```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

```python

X, y = make_blobs(n_samples=300, centers=3, random_state=42, cluster_std=1.0)
```

```python
plt.scatter(X[:, 0:1], X[:, 1:2], s=20)
plt.title("Generated Data Points")
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
```

```python
kmeans.fit(X)
```

```python
centroids = kmeans.cluster_centers_
labels = kmeans.labels_
```

```python
plt.scatter(X[:, 0], X[:, 1], c=labels, s=20, cmap='viridis')
```

```python
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=90, marker='X')
plt.title("K-Means Clustered Data")
plt.show()
```

```python
print(f"Inertia: {kmeans.inertia_}")

```
